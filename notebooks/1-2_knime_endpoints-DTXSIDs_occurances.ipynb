```python
#%reset
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
endpoint_data = pd.read_csv("../data/tox21-data/endpoints_oer_DTXSID.csv")
endpoint_data["Sum(hitc)"].describe()
endpoint_data.describe()
endpoint_data["hit_perc"] = endpoint_data["Sum(hitc)"]/endpoint_data["Count*(hitc)"]*100
endpoint_data.head()
endpoint_data.describe()
DTXSID_data = pd.read_csv("../data/tox21-data/DTXSID_oer_endpoint.csv")
DTXSID_data.head()
```

```python
sns.set(font_scale = 3)
sns.set_style("white")
rc = {'figure.figsize':(10,10),
      'axes.facecolor':'white',
      'axes.grid' : False,
      'grid.color': '.8',
      #'font.family':'Times New Roman',
      'font.size' : 25}
plt.rcParams.update(rc)

ax = sns.histplot(log_scale=False, edgecolor=".3", 
             linewidth=.5, data=endpoint_data, x="Count*(hitc)", palette = "BrBG")
sns.despine(fig=None, ax=None, top=False, right=False, left=False, bottom=False, offset=None, trim=False)
ax.set(xlabel='# chemicals', ylabel='# endpoints')
ax.set_xlim(0,10000)
ax.set_title("Endpoints per chemicals")
sns.set_context("paper", font_scale=2) 
plt.savefig('../plots/tox21/chemicals_per_endpoints.svg')
```

```python
sns.set(font_scale = 3)
sns.set_style("white")
rc = {'figure.figsize':(10,10),
      'axes.facecolor':'white',
      'axes.grid' : False,
      'grid.color': '.8',
     #'font.family':'Times New Roman',
      'font.size' : 15}
plt.rcParams.update(rc)

ax = sns.histplot(log_scale=False, edgecolor=".3", 
             linewidth=.5, data=DTXSID_data, x="Count*(hitc)", palette = "BrBG")
sns.despine(fig=None, ax=None, top=False, right=False, left=False, bottom=False, offset=None, trim=False)
ax.set(xlabel='# endpoints', ylabel='# chemicals')
ax.set_xlim(0,1000)
ax.set_title("Chemicals per endpoints")
plt.savefig('../plots/tox21/endpoints_per_chemicals.svg')
```
