```python
plt.figure(figsize=(16,6))
sns.set()
rdgn = sns.diverging_palette(h_neg=130, h_pos=10, s=99, l=55, sep=3,as_cmap=True)
cmap = sns.cm.rocket_r
yticks = df2.dtxsid
df3 = df2.drop(['dtxsid'], axis=1)
keptticks = yticks[::int(len(yticks)/20)]
yticks = ['' for y in yticks]
yticks[::int(len(yticks)/20)] = keptticks
xticks = df3.columns
keptticks = xticks[::int(len(xticks)/20)]
xticks = ['' for y in xticks]
xticks[::int(len(xticks)/20)] = keptticks
ax = sns.heatmap(df3, cmap = "gray", cbar=False)
# , yticklabels=False, xticklabels=False
for _, spine in ax.spines.items():spine.set_visible(True)
ax.set(xlabel='Fingerprint bits', ylabel='MassBank chemicals (#index)')
plt.savefig('../plots/massbank/MolFps_truevspred_heatmap2.png', bbox_inches = 'tight')
plt.show()
```
