```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

dfBarPlot = pd.read_csv('truepred_molfps_perf_mechanistic_targets.csv')
#duplicate observations to get good std bars
dfCopy = dfBarPlot.copy()
duplicates = 30 # increase this number to increase precision
for index, row in dfBarPlot.iterrows():
    for times in range(duplicates):
        new_row = row.copy()
        new_row['recall'] = np.random.normal(row['recall'],row['recall_dev']) 
        dfCopy = dfCopy.append(new_row, ignore_index=True)

# Now Seaborn does the rest
sns.set_style("whitegrid")
fig = sns.barplot(x='mechanistic',
                  y='recall',
                  hue='origin',
                  ci='sd',
                  data=dfCopy)
#003c30
#177b73
#75c5ba
#d4edea
#f6ecd1
#dab972
#a26619
#543005

colors = ["#003c30", "#dab972"]
sns.set_palette(sns.color_palette(colors))
plt.legend(loc='upper right')
sns.set(rc={'figure.figsize':(24,12)})
plt.show()
```
